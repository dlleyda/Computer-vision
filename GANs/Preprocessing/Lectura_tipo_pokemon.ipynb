```python
import json

with open('pokemon.json', encoding='utf-8') as file:
    data = json.load(file)
```

```python
result_dict = {}

for entry in data:
    for type_ in entry["type"]:
        if type_ not in result_dict:
            result_dict[type_] = [entry["id"]]
        else:
            result_dict[type_].append(entry["id"])
```

```python
modified_data = []

for entry in data:
    modified_entry = entry.copy()
    modified_entry.pop("base", None)
    modified_entry.pop("name", None)
    modified_data.append(modified_entry)
```

```python
excepciones = {}
```

```python
import os
import re

folder_path = './pokemon/main-sprites/black-white'  # Replace with the actual path to your folder


png_files = []

# Check if the folder exists
if os.path.exists(folder_path) and os.path.isdir(folder_path):
    # Get a list of all files in the folder
    all_files = os.listdir(folder_path)

    # Filter files with .png extension
    png_files = [file for file in all_files if file.lower().endswith('.png')]

else:
    print(f"The specified folder path '{folder_path}' does not exist or is not a directory.")
    
file_dict = {}

for file_name in png_files:
    # Use regular expression to extract the numeric part
    match = re.search(r'(\d+)', file_name)
    
    if match:
        numeric_part = int(match.group(1))
        file_dict[file_name] = numeric_part
        
    if file_name in excepciones.keys():
        file_dict[file_name] = file_name

print(file_dict)
```

```python
diccionario_resultante = {}

for key, values in result_dict.items():
    diccionario_resultante[key] = [next((filename for filename, value in file_dict.items() if value == v), None) for v in values]

print(diccionario_resultante)
```
