```python
from PIL import Image

def upscale_image(input_path, output_path):
    # Open the image
    original_image = Image.open(input_path)

    # Upscale using bicubic interpolation
    upscaled_image = original_image.resize((256, 256), Image.BICUBIC)

    # Save the upscaled image
    upscaled_image.save(output_path)

# Example usage:
# input_image_path = "generation-5/pokemon/main-sprites/black-white/1.png"
# output_image_path = "C:/Users/David/Desktop/U-TAD/5to AÃ‘O/1er cuatrimestre/Vision por computador/Practicas/Practica 3, Tema 4/upscaled_image.png"

upscale_image(input_image_path, output_image_path)
```
